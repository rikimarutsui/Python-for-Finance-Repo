```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('time_data/walmart_stock.csv', index_col='Date', parse_dates=True)
```

```python
df.head()
```

```python
df['Open'].plot(figsize=(16,6))
```

```python
df.rolling(7).mean().head(14)
# Rolling 7 days group by mean and get first 14 rows
# Becasue first 6 rows don't have enough data to calculate (fully 7 days)
```

```python
df['Open'].plot()
df.rolling(window=7).mean()['Close'].plot(figsize=(16,6))
# Show Open in Blue and Close in Orange
```

```python
df['Open'].plot()
df.rolling(window=30).mean()['Close'].plot(figsize=(16,6))
# Show Open in 7 days, Close in 30 days
```

```python
df['Open'].plot()
df.rolling(window=90).mean()['Close'].plot(figsize=(16,6))
# Show Open in 7 days, Close in 90 days
```

```python
df['Close 30 Day MA'] = df['Close'].rolling(window=30).mean()
df[['Close 30 Day MA', 'Close']].plot(figsize=(16,6))
```

```python
# Expanding
df['Close'].expanding().mean().plot(figsize=(16,6))
```

```python
# Close 20 MA
df['Close: 20 Day Mean'] = df['Close'].rolling(window=20).mean()

# Upper = 20MA + 2*std(20)
df['Upper'] = df['Close: 20 Day Mean'] + 2 * (df['Close'].rolling(20).std())

# Lower = 20MA - 2*std(20)
df['Lower'] = df['Close: 20 Day Mean'] - 2 * (df['Close'].rolling(20).std())

# Close
df[['Close', 'Close: 20 Day Mean', 'Upper', 'Lower']].plot(figsize=(16,6))
```
