```python
fig, ax = plt.subplots()
ax.plot_date(idx, stock)
ax.plot_date(idx, stock, '-')

# LOCATING Month
ax.xaxis.set_major_locator(dates.MonthLocator())
ax.xaxis.set_major_formatter(dates.DateFormatter('%b-%Y'))

fig.autofmt_xdate()
plt.tight_layout()
```

```python
fig, ax = plt.subplots()
ax.plot_date(idx, stock)
ax.plot_date(idx, stock, '-')

# LOCATING Month
ax.xaxis.set_major_locator(dates.MonthLocator())
ax.xaxis.set_major_formatter(dates.DateFormatter('%b/%Y'))

fig.autofmt_xdate()
plt.tight_layout()
```

```python
fig, ax = plt.subplots()
ax.plot_date(idx, stock)
ax.plot_date(idx, stock, '-')

# LOCATING Month \n -- next line
ax.xaxis.set_major_locator(dates.MonthLocator())

# FORMATING
ax.xaxis.set_major_formatter(dates.DateFormatter('\n\n%b/%Y'))

fig.autofmt_xdate() # Autoformat the date in X to easy readable
plt.tight_layout()
```

```python
fig, ax = plt.subplots()
ax.plot_date(idx, stock)
ax.plot_date(idx, stock, '-')

# LOCATING Month \n -- next line
# MonthLocator 0 = jan, 1 = feb ...... 11 = dec
ax.xaxis.set_major_locator(dates.MonthLocator())
# FORMATING
ax.xaxis.set_major_formatter(dates.DateFormatter('\n\n\n\n%b-%Y'))

# WeekdayLocator 0 = mon, 1 = tue ...... 6 = sun
ax.xaxis.set_minor_locator(dates.WeekdayLocator(byweekday=0))
ax.xaxis.set_minor_formatter(dates.DateFormatter('%a')) # %a is the weekday

fig.autofmt_xdate() # Autoformat the date in X to easy readable
plt.tight_layout()
```

```python
fig, ax = plt.subplots()
ax.plot_date(idx, stock)
ax.plot_date(idx, stock, '-')

# LOCATING Month \n -- next line
# MonthLocator 0 = jan, 1 = feb ...... 11 = dec
ax.xaxis.set_major_locator(dates.MonthLocator())
# FORMATING
ax.xaxis.set_major_formatter(dates.DateFormatter('\n\n\n\n%b-%Y'))

# WeekdayLocator 0 = mon, 1 = tue ...... 6 = sun
ax.xaxis.set_minor_locator(dates.WeekdayLocator(byweekday=0))
ax.xaxis.set_minor_formatter(dates.DateFormatter('%d')) # %d is actual day

fig.autofmt_xdate() # Autoformat the date in X to easy readable
plt.tight_layout()
```
